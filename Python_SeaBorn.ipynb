```python
# Python seaborn library 

# Line Plot
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
fmri=sns.load_dataset("fmri")
fmri.head()
```

```python
sns.lineplot(x='timepoint', y='signal', data=fmri)
plt.show()
```

```python
# hue parameter
```

```python
sns.lineplot(x='timepoint', y='signal', data=fmri, hue='event')
plt.show()
```

```python
# adding styles
```

```python
sns.lineplot(x='timepoint', y='signal', data=fmri, hue='event', style='event')
plt.show()
```

```python
# adding markers
```

```python
sns.lineplot(x='timepoint', y='signal', hue='event', style='event', markers=True, data=fmri)
plt.show()
```

```python
import pandas as pd
```

```python
sns.boxplot(x='InternetService', y='MonthlyCharges', data=churn)
plt.show()
```

```python
sns.boxplot(x='Contract', y='tenure', data=churn, palette='Set1')
plt.show()
```

```python
sns.boxplot(x='Contract', y='tenure', data=churn, linewidth=3)
plt.show()
```

```python
sns.barplot(x='Contract', y='tenure',data=churn, order=["Two year", "Month-to-month", "One year"])
plt.show()
```

```python
sns.boxplot(x='Contract', y='tenure', data=churn, hue='PaymentMethod')
plt.show()
```
