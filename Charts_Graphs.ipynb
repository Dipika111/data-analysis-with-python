```python
x=np.arange(1,11)
y1=2*x
y2=3*x

plt.subplot(2,1,1)
plt.plot(x,y1,color='r',linestyle=':')

plt.subplot(2,1,2)
plt.plot(x,y2,color='y',linewidth=4)

plt.title("Two Subplots")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(True)
plt.show()
```

```python
# Scatter plot
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40,50,60,70,80,90]
y=[8,1,7,2,0,3,7,3,2]
```

```python
plt.title("scatter plot")
plt.scatter(x,y,marker="*",c="r",s=100)
plt.show()
```

```python
a=[10,20,30,40,50,60,70,80,90]
b1=[8,1,7,2,0,3,7,3,2]
b2=[7,2,5,6,9,1,4,5,3]
```

```python
plt.title("subplot scatter")
plt.subplot(1,2,1)
plt.scatter(a,b1,marker="*",c="g",s=100)


plt.subplot(1,2,2)
plt.scatter(a,b2,marker=".",c="r",s=200)
plt.show()
```

```python
# Pie- chart
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
fruit=["apple","kiwi","mango","orenge","banana"]
quantity=[10,5,15,20,8]
```

```python
plt.pie(quantity,labels=fruit,autopct='%0.1f%%',colors=["yellow","blue","green","red","black"])
plt.show()
```

```python
# Donut chart
```

```python
fruit=["apple","kiwi","mango","orenge","banana"]
quantity=[10,5,15,20,8]
```

```python
plt.pie(quantity,labels=fruit,radius=2)
plt.pie([1],colors=['w'],radius=1)
plt.show()
```
